Plotting the gamma gated beta times produced by James.

## Neutron gated asymmetry

data from James gated on the strongest neutron peak.

```julia
ndata = CSV.read("49K_KCl_T005_laser_on_ntof.csv",DataFrame)
```

```julia
display(plot(ndata[:,2],seriestype=:stephist,xlims=(150,500),nbins=3050))

plot(ndata[ndata[:,1].==3,3],seriestype=:stephist,xlims=(0,400),nbins=5001)

beta1=fit(Histogram,ndata[ndata[:,1].==2,3],0:40:3500);
beta2=fit(Histogram,ndata[ndata[:,1].==3,3],0:40:3500);

asymmetry = (beta1.weights .- beta2.weights) ./ (beta1.weights + beta2.weights)

plot(eachindex(asymmetry).*40,asymmetry,seriestype=:line,markersize=4)
```

let's try to slice the asymmetry by neutron ToF

```julia
step = 4
ntof = fit(Histogram,ndata[:,2],150:1:300)
pbeta1=[]
pbeta2=[]
for i in range(150,300,step=step)    
    push!(pbeta1,length(ndata[ndata[:,1].==2 .&& (ndata[:,2] .> i .&& ndata[:,2] .< i+step),3]))
    push!(pbeta2,length(ndata[ndata[:,1].==3 .&& (ndata[:,2] .> i .&& ndata[:,2] .< i+step),3]))
end

pasymmetry = (pbeta1.-pbeta2) ./ (pbeta1.+pbeta2)

display(plot(
            plot(ntof,seriestype=:step,xlabel="ToF (ns)",label="LASER ON"),

            plot(range(150,300,step=step),pasymmetry,seriestype=:scatter,
            yerror = (pasymmetry.* 2 .*sqrt.(pbeta1)./pbeta1),
            xlabel = "ToF (ns)", ylabel = "Aᵦ",label =""),

            layout = @layout [a;b]))
savefig("ntof_asymmetry.pdf")
```

Now for the fun part we redo the plot for the laser off file

```julia
nolaserdata = CSV.read("49K_KCl_T006_laser_off_ntof.csv",DataFrame)
```

```julia
step = 4
ntof = fit(Histogram,nolaserdata[:,2],150:1:300)
plaseroffbeta1=[]
plaseroffbeta2=[]
for i in range(150,300,step=step)    
    push!(plaseroffbeta1,length(nolaserdata[nolaserdata[:,1].==2 .&& (nolaserdata[:,2] .> i .&& nolaserdata[:,2] .< i+step),3]))
    push!(plaseroffbeta2,length(nolaserdata[nolaserdata[:,1].==3 .&& (nolaserdata[:,2] .> i .&& nolaserdata[:,2] .< i+step),3]))
end

plaseroffasymmetry = (plaseroffbeta1.-plaseroffbeta2) ./ (plaseroffbeta1.+plaseroffbeta2)

display(plot(
            plot(ntof,seriestype=:step,xlabel="ToF (ns)",label="LASER OFF"),

            plot(range(150,300,step=step),plaseroffasymmetry,seriestype=:scatter,
            yerror = (plaseroffasymmetry.* 2 .*sqrt.(plaseroffbeta1)./plaseroffbeta1),
            xlabel = "ToF (ns)", ylabel = "Aᵦ",label =""),

            layout = @layout [a;b]))
savefig("ntof_nolaser_asymmetry.pdf")
```
