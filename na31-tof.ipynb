# Na-31 ToF from the INDiE campaign October 2023

We plot the 31Na using Xu's dataset.

```julia
using DelimitedFiles,Plots,Plots.PlotMeasures


```

```julia
default(fontfamily="Computer Modern Bold",  
        color=:navy,linewidth=2, framestyle=:box, 
        label=nothing, grid=false,
        xlabel = "Eáµ§ (keV)")
default(guidefont=(14, :black),tickfont=(12,:black))

tof(x) = sqrt(0.5*939.565*100^2/(30^2*x))
```

```julia
data = readdlm("31Na_INDIE_tof.txt")

naplot=plot(data[:,1],data[:,2],xlims=(20,390),ylims=(0,1350),xlabel="ToF (ns)",dpi=400,
ylabel="counts/250 ps",legend=false,top_margin=5mm,framestyle=:semi)

energies=[ range(0.2,0.9,step=0.1); range(2.0,6.0,step=1.0) ]
majorenergies =[0.1,1]
for E in energies plot!([tof(E),tof(E)],[1300,1350],color=:black) end
for E in majorenergies plot!([tof(E),tof(E)],[1290,1350],color=:black) end


# vline!([257])

annotate!(tof(0.1),1260,"0.1")
annotate!(tof(1.0),1260,"1.0")
annotate!(200,1410,"E (MeV)")

savefig("na31tof.png")
display(naplot)
```
