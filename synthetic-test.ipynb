```julia
data=[100  10
      101  10
      102  10
      103  10
      104  100
      105  400
      106  120
      107  10
      108  10
      109  10
      110  10
      111  10
      ]

println(sum(data[:,2])-10*length(data[:,1]))

plot(data[:,1],data[:,2],seriestype=:stepmid,lc=:navy,lw=1.5)
```
